```python
#  Project: "Password Strength Analyzer & Generator"
# Objective:
# Build a tool that:

# Generates random passwords using NumPy and strings.

# Analyzes password strength based on rules like length, diversity of characters, and entropy.
```

```python
import string

lowercase = list(string.ascii_lowercase)
uppercase = list(string.ascii_uppercase)
digits = list(string.digits)
symbols = list(string.punctuation)
```

```python
import numpy as np

def generate_password(length=12):
    all_chars = lowercase + uppercase + digits + symbols
    password = np.random.choice(all_chars, size=length)
    return ''.join(password)
```

```python
print(generate_password(8))
print(generate_password(16))
```

```python
def password_strength(password):
    score = len(password)
    if any(c.islower() for c in password) and any(c.isupper() for c in password):
        score += 5
    if any(c.isdigit() for c in password):
        score += 5
    if any(c in symbols for c in password):
        score += 5
    if len(password) > 12:
        score += 5
    return score
```

```python
lengths = np.arange(6, 40)
strengths = [password_strength(generate_password(length)) for length in lengths]

import matplotlib.pyplot as plt
plt.plot(lengths, strengths, marker='o')
plt.title("Password Strength vs Length")
plt.xlabel("Password Length")
plt.ylabel("Strength Score")
plt.grid(True)
plt.show()
```
